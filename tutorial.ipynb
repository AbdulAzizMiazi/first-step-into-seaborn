```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# %matplotlib inline
# %reload_ext autoreload
# %autoreload 2
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

**Destribution Plot**

```python
# sns.distplot() <= this function is going to be removed from the future versions of Seaborn. New is sns.displot()
sns.displot(crash_df['not_distracted'], bins=10);
```

**Joint Plot**

```python
# Regration
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg');
# kernel density estimate (KDE) Plot
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde');
# Hexagon Ploat
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex');
```

**Pair Plots**

```python
# sns.pairplot(crash_df)
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

**Rug Plot**

```python
sns.rugplot(tips_df['tip'])
```
